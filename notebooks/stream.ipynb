```python
import plasma.data_model.collections as pdc
```

```python
a = pdc.Stream(range(10)).select(str).filter(lambda s: int(s) < 5).unwind(lambda x:x)
a.evaluate()
```

```python
pdc.Stream(range(10)).enumerate().select(lambda i, j: (i, str(j))).evaluate()
```

```python
pdc.Stream(range(10)).enumerate().project(lambda i, j: i * j).evaluate()
```

```python
pdc.Stream(range(10)).groupby(lambda i: i % 2, lambda i: i // 2).filter(lambda k, vs: k == 0).evaluate()
```

```python
pdc.Stream(range(10)).take(5).evaluate()
```
