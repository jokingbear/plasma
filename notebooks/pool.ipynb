```python
import time
import itertools

from plasma.parallel_processing import pool
```

```python
def compute(x:int):
    time.sleep(0.5)
    return x + 1
```

```python
stream = pool.Process(5).stream(range(20))\
                        .select(compute).filter(lambda a: a < 12)\
                        .unwind(lambda a: itertools.product([a], range(20)))\
                        .groupby(lambda a:a[0], lambda a:a[1])\
                        .select(print)
```

```python
stream.to_list()
```
