```python
import plasma.functional as F
```

```python
def test(x:int): return x + 1
```

```python
def add(x, y): return x + y
```

```python
f = (
    F.pipe(test)
    .chain(test).chain(test).chain(test)
    .partial_left(add, 1)
    .chain(lambda x: f'hehe: {x}')
)
f
```

```python
f(8)
```
