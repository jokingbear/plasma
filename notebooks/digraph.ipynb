```python
from plasma.data_model.graph import DiGraph
from plasma.data_model import model, dataclass
```

```python
@model
class A:
    a1:int
    a2:int
```

```python
a = DiGraph(A.a1)
```

```python
data = [
    A(8, 9),
    A(9, 10),
    A(10, 11),
]

a.add_nodes_from((i, {'type': A, 'data': d}) for i, d in enumerate(data))
a.add_edges_from([(0, 1), (1, 2)])
```

```python
[*a.inquirer.nodes(A)], \
    [*a.inquirer.nodes(A).select('a1', A.a1, a2=A.a2, a3=lambda _,__,td:td.a2+5, a4='a1')]
```
