```python
import h5py as h5
import imageio as io
```

```python
from glob import glob
```

```python
files = list(glob("test/*.png"))

len(files)
```

```python
h5_file = h5.File("test.h5", mode="w")

h5_file
```

```python
img_db = h5_file.create_dataset("image", shape=[len(files), 256, 256])
```

```python
from tqdm import tqdm_notebook as tqdm
```

```python
for i, f in tqdm(enumerate(files), total=len(files)):
    img = io.imread(f) / 127.5 - 1
    img_db[i] = img
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img_db[0])
plt.show()
```

```python
h5_file.close()
```
